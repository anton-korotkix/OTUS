# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
one_d_array = np.zeros(10) #создаю одномерный массив из 10 нулей
one_d_array[4] = 1 #пятый элемент приравниваю 1
two_d_array = one_d_array.reshape((5, 2)) #трансформирую в двумерный массив
print(two_d_array) #вывод
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
one_d_array = np.arange(10, 50) #создаю одномерный массив от 10 до 49
one_d_array_rev = one_d_array[::-1] #разворачиваю
result = one_d_array_rev[one_d_array_rev % 2 == 0] #ищу все четные элементы
print(result) #вывод
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
arr = np.arange(9).reshape(3, 3) #создаю массив и трансформирую в 3 на 3
print(arr) #вывод
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.rand(4, 3, 2) #создаю массив 4 на 3 на 2
print(arr) #вывод массива
min_element = np.min(arr) #ищу минимальный элемент
max_element = np.max(arr) #ищу максимальный элемент
print(min_element, max_element) #вывод мин и макс
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение.

```python
arr1 = np.random.rand(6, 4) #создаю массив 6 на 4
arr2 = np.random.rand(4, 3) #создаю массив 4 на 3
arr = np.matmul(arr1, arr2)
print(arr) #вывод массива
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset('tips') #загрузил датасет
```

### Посмотрите на первые 5 строчек

```python
print(tips.head()) #вывод первых 5 строк
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(tips.shape) #вывод общего количества строк и столбцов
```

### Проверьте есть ли пропуски в данных

```python
tips.isnull().sum() #вывод количества пропусков в данных по столбцам
```

### Посмотрите на распределение числовых признаков

```python
tips.describe() #распределение числовых признаков
```

### Найдите максимальное значение 'total_bill'

```python
max_total_bill = tips['total_bill'].max() #поиск максимума
print(max_total_bill) #вывод
```

### Найдите количество курящих людей

```python
result = (tips['smoker'] == 'Yes').sum() #суммарное количество курящих
print(result) #вывод
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
mean_day_total_bill = tips.groupby(['day'])['total_bill'].mean() #ищу среднее по дням
print(mean_day_total_bill) #вывод
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_total_bill = tips['total_bill'].median() #ищу медиану
big_median_total_bill = tips[tips['total_bill'] > median_total_bill] #ищу которые больше медианы
mean_tip = big_median_total_bill.groupby('sex')['tip'].mean() #ищу средний tip с группировкой по полу
print(mean_tip) #вывод
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1}) #преобразование
```
